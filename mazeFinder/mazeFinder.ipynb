```python
import numpy as np
import os
```

```python
maze = []

file = open('maze.txt','r')

for line in file:
    numString = list(line)
    maze.append(numString)
```

```python
stack = []

found = False
```

```python
def find_start():
    for i in range(len(maze)):
        for j in range(len(maze)):
            if (maze[i][j] == "S"):
                stack.append((i,j))

def is_taken(cord):
    global found
    if (0 <= cord[0] < len(maze)) & (0 <= cord[1] < len(maze)):
        if maze[cord[0]][cord[1]] == 'E':
            found = True
            print(cord[0] + cord[1])
            print('found')
        if maze[cord[0]][cord[1]] == '_':
            maze[cord[0]][cord[1]] = '-'
            return False
    return True

def checkAround(cord):
    global stack, found
    cords = []
    cords.append((cord[0] + 1,cord[1]))
    cords.append((cord[0] - 1,cord[1]))
    cords.append((cord[0],cord[1] + 1))
    cords.append((cord[0],cord[1] - 1))
    
    for card in cords:
        if not found:
            print(card)
            v = is_taken(card)
            print(v)
            if not v:
                stack.append(card)
            
def mainLoop():
    global stack
    print(stack)
    for i in stack:
        temp = stack.pop()
        checkAround(temp)
        
        
    
```

```python
find_start()

while not found:
    mainLoop()

    
```
